```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


car_dataset= pd.read_csv('C:/Users/vaish/OneDrive/Desktop/docs/jupyter/car_ad.csv', sep=',', encoding='latin-1')

from sklearn.linear_model import LinearRegression
X = car_dataset['mileage'].values.reshape(-1,1)
Y= car_dataset['price'].values.reshape(-1,1)
reg= LinearRegression()
reg.fit(X,Y)

print(reg.coef_[0][0])
print(reg.intercept_[0])

predictions= reg.predict(X)
plt.figure(figsize= (16,8))
plt.scatter( car_dataset['mileage'],car_dataset['price'], color= 'red')
plt.plot(car_dataset['mileage'], predictions, color='black')
plt.xlabel("mileage")
plt.ylabel("price")
plt.show()
```

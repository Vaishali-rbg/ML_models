```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
housing_data= pd.read_csv('C:/Users/vaish/OneDrive/Desktop/docs/jupyter/Housing.csv' , sep=',', encoding='latin-1')

plt.figure(figsize= (16,8))
plt.scatter( housing_data['area'],housing_data['price'])
plt.xlabel("area")
plt.ylabel("price")
plt.show()
```

```python
import seaborn as sns
sns.lmplot(x ='area', y = 'price', data = housing_data)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
X = housing_data['area'].values.reshape(-1,1)
Y= housing_data['price'].values.reshape(-1,1)
reg= LinearRegression()
reg.fit(X,Y)

print(reg.coef_[0][0])
print(reg.intercept_[0])

predictions= reg.predict(X)
plt.figure(figsize= (16,8))
plt.scatter( housing_data['area'],housing_data['price'])
plt.plot(housing_data['area'], predictions)
plt.xlabel("area")
plt.ylabel("price")
plt.show()
```
